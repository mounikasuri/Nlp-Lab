**Removing Noise** **Data**

```python
noise_list=["is","a","the","..."]
def remove_noise(input_text):
  words=input_text.split()
  noise_free_words=[word for word in words if word not in noise_list]
  noise_free_text=" ".join(noise_free_words)
  return noise_free_text
remove_noise("this is a sample")
```

**Remove Regular Expression pattern**

```python
import pandas as pd
data = pd.read_csv("/content/Coachella-2015-2-DFE.csv")
```

```python

import re
def extract_hashtag_words(text):
    hashtags = re.findall(r'#(\w+)', text)
    return hashtags
hashtag_words_list = data['text'].apply(extract_hashtag_words)
all_hashtag_words = [word for sublist in hashtag_words_list for word in sublist]
unique_hashtag_words = set(all_hashtag_words)
for word in unique_hashtag_words:
    print(word)
```

```python

import pandas as pd
import re

def remove_hashtags(text):
    return re.sub(r'#\w+', '', text)
data['text_without_hashtags'] = data['text'].apply(remove_hashtags)
print (data[['text_without_hashtags']].head())
```
